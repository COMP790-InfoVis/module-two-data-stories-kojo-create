```python
import pandas as pd
df = pd.read_csv('all_seasons.csv')
df.head()
```

```python
specific_players = ['LeBron James', 'Kevin Durant', 'Luka Doncic']
filtered_df = df[df['player_name'].isin(specific_players)]
```

```python
import altair as alt
# Create the scatterplot using the filtered DataFrame
scatter_plot = alt.Chart(filtered_df).mark_circle(size=100).encode(
    x='season',
    y='pts',  
    color='player_name',  # Assures that different players will have different colors
    tooltip=['player_name', 'pts', 'season']  # Shows details on hover
).properties(
    width=600,
    height=400,
    title='Specific NBA Players: Scoring Efficiency vs. Career Longevity'
)

# Display the plot
scatter_plot
```
